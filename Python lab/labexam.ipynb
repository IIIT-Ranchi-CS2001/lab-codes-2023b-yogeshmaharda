```python
import pandas as pd

# Load the dataset
df = pd.read_csv("./AQI_Data.csv")

print("First 8 rows: ")
print(df.head(8))
```

```python
print("Last 5 rows: ")
print(df.tail(5))
```

```python
# dtype of each column
print("Data types of each column: ")
print(df.dtypes)

# Number of non null values
print(f"\nNumber of non null values:\n {df.count()}")
```

```python
import numpy as np

# Calculate statistical parameters using numpy
meanAQI = np.mean(df["AQI"])
maxPM2dot5 = np.max(df["PM2.5"])
minPM10 = np.min(df["PM10"])

print(f"Mean AQI: {meanAQI}")
print(f"Maximum PM2.5 {maxPM2dot5}")
print(f"Minimum PM10: {minPM10}")
```

```python
# Rows corresponding to unique cities

cities = np.unique(df["City"], return_counts=True)
cityNames = cities[0]
numRows = cities[1]

citycount = dict()
# put the data in a dictionary
for i in range(0, len(cityNames)):
    citycount[cityNames[i]] = numRows[i]
    
print(citycount)
```

```python
# Calculate sum of pollutants
df["Pollutant Sum"] = np.sum(df[["PM2.5", "PM10", "NO2", "CO", "O3", "SO2"]], axis=1)

# Display the Dataframe with Pollutant Sum column
print(df.head())

# 
```

```python
# save the file
try:
    df.to_csv("pollutant.csv")
except PermissionError:
    print("Could not save file: Restricted by system")
except FileNotFoundError:
    print("New file could not be craeted")
except:
    print("An unknown Error occured while saving data")
```
